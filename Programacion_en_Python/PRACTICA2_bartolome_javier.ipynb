#### 1

Diseña una función llamada `municipios` cuyo dato de entrada sea el contenido del fichero y la salida sea la lista de municipios contenidos en el fichero. ¿Cuantos municipios hay? 

```python
#Mediante una función while creamos una lista vacia para ir añadinedo el numero de municipios y el respectivo nombre de cada uno
# mas tarde en la funcion le pedimos que nos diga que tamaño tiene la lista
def municipios(datos):
    lista = [] 
    j = 0 
    lista_datos = datos['data'] 
    while j < len(lista_datos):
        nombre_municipio = lista_datos[j]['municipio_nombre'] 
        if nombre_municipio in lista: 
            j = j + 1
        else: 
            j = j + 1
            lista.append(nombre_municipio)
    print(len(lista)) 
    return lista
```

```python
municipios(datos)
```

Diseña una función llamada `nacimientos` cuyo dato de entrada sea el contenido del fichero y la salida sea el número total de nacimientos registrados en el fichero. 

```python
#Mediante una función while creamos una lista vacia2 para ir añadinedo el numero de nacimientos
# mas tarde en la funcion le pedimos que nos diga que la suma del recuento

def nacimientos (datos): 
    lista2 = datos['data']
    a = 0 
    lista_nacimientos = [] 
    while a < len(lista2):
        nacim = lista2[a]['numero_nacimientos']
        a = a + 1
        lista_nacimientos.append(nacim)
    return sum(lista_nacimientos)  
nacimientos(datos) 

```
